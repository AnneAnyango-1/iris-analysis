```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Iris dataset from seaborn (built-in)
df = sns.load_dataset('iris')
# Showing the first 5 rows
print(df.head())
```

```python
# Checking data types and missing values
print(df.info())
print(df.isnull().sum())
```

```python
# Cleaning data: drop rows with missing values (if any)
df = df.dropna()
```

```python
# Summarizing statistics
print(df.describe())
```

```python
# Grouping by species and compute mean of numerical columns
grouped = df.groupby('species').mean()
print(grouped)
```

```python
# Observations
print("Observation: Setosa species has smallest average petal length.")
```

```python
# 1. Line chart 
df['index'] = df.index
df.sort_values('index').groupby('species')['sepal_length'].plot(legend=True)
plt.title("Sepal Length Trend by Species")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
# 2. Bar chart
df.groupby('species')['petal_length'].mean().plot(kind='bar')
plt.title("Average Petal Length by Species")
plt.ylabel("Petal Length")
plt.show()
```

```python
# 3. Histogram
df['sepal_width'].plot(kind='hist', bins=10)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.show()
```

```python
# 4. Scatter plot
df.plot(kind='scatter', x='sepal_length', y='petal_length', c='blue')
plt.title("Sepal Length vs Petal Length")
plt.show()
```
